```python
from tensorflow.python.tools.inspect_checkpoint import print_tensors_in_checkpoint_file
import os

model_dir = 'checkpoints/'
checkpoint_path = os.path.join(model_dir, "model.ckpt-2961")

# print all tensors in checkpoint file
print_tensors_in_checkpoint_file(file_name=checkpoint_path, tensor_name='', all_tensors=True)
```

```python
# print out all the tensors name
print_tensors_in_checkpoint_file(file_name=checkpoint_path, tensor_name='', all_tensors=False, all_tensor_names=True)
```

```python
# print only tensor "tensor_name" in ckeck point file
print_tensors_in_checkpoint_file(file_name=checkpoint_path, tensor_name='w_conv_1', all_tensors=False)
```
