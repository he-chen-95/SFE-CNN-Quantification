#### original_shape=(5 * 5* 1* 16)

#### current_shape=(5 columns * 80 rows)

every 5 row represent a 2D convolutio kernal

#### target_shape=(25 * 16)

each rows represent a 2D convolution kernel

```python
# df1 include [0, 5, 10, 15 ... 75]
# df2 include [1, 6, 11, 16 ... 76]
# df3 include [2, 7, 12, 17 ... 77]
# df4 include [3, 8, 13, 18 ... 78]
# df5 include [4, 9, 14, 19 ... 79]

# Horizontally connect file 1, file 2, file 3
ranges_df0 = range(0, 80, 5)
ranges_df1 = range(1, 80, 5)
ranges_df2 = range(2, 80, 5)
ranges_df3 = range(3, 80, 5)
ranges_df4 = range(4, 80, 5)

#print("ranges_df0:\n", ranges_df0)
#print("ranges_df1:\n", ranges_df1)
#print("ranges_df2:\n", ranges_df2)
#print("ranges_df3:\n", ranges_df3)
#print("ranges_df4:\n", ranges_df4)

# The first parameter is used to select the row
# The second parameter is used to select the column
# df0 = w_conv_2.iloc[ranges_df0, ['0', '1', '2', '3', '4']]
df0 = w_conv_1.iloc[ranges_df0]
df1 = w_conv_1.iloc[ranges_df1]
df2 = w_conv_1.iloc[ranges_df2]
df3 = w_conv_1.iloc[ranges_df3]
df4 = w_conv_1.iloc[ranges_df4]
print(df0.head())
print(df1.head())
print(df2.head())
print(df3.head())
print(df4.head())
```

```python
# Index reordering
df0.index = range(len(df0))
df1.index = range(len(df1))
df2.index = range(len(df2))
df3.index = range(len(df3))
df4.index = range(len(df4))
print(df0.head())
print(df1.head())
print(df2.head())
print(df3.head())
print(df4.head())
```

```python
# Splicing files horizontally
w_conv_1_v1 = pd.concat([df0, df1, df2, df3, df4], axis=1)
w_conv_1_v1.head()
```

```python
# change the columns name to 0->24
# print(w_conv_1_v1.columns.values)
# Edit specific columns
#w_conv_1_v1.rename(columns={w_conv_1_v1.columns[0]: "new_name"}, inplace=True)
ranges_columns_name_v1 = range(0, 25, 1)
w_conv_1_v1.columns = ranges_columns_name_v1
print("number of rows:", w_conv_1_v1.index)
w_conv_1_v1.tail()
```
